### Sales
- **ID_SALES**: ID sales department
- **Program_Code**: Program code of the trip
- **From_Grade**: Lowest grade in school of a participant
- **To_Grade**: Highest grade in school of a participant
- **Group_State**: School location
- **Days**: Number of days on the program
- **Travel_Type**: Travel mode (A = Air, B = Bus, T = Train)
- **Departure_Date**: Departure day
- **Return_Date**: Return day
- **Early_RPL**: First communication date inviting people to join
- **Latest_RPL**: Last communication inviting people to join
- **Cancelled_Pax**: Number of passengers who made a deposit but cancelled
- **Total_Discount_Pax**: Number of extra passengers (e.g. professors)
- **Initial_System_Date**: First date when trip was organized
- **SPR_Product_Type**: Aggregation of tour types
- **FPP**: Number of full-payment participant
- **Total_Pax**: Number of total passengers (including extra participants)
- **DepartureMonth**: Month of departure
- **GroupGradeTypeLow**: Lowest grade type in the trip
- **GroupGradeTypeHigh**: Highest grade type in the trip
- **GroupGradeType**: Combination of the above
- **MajorProgramCode**: Aggregation of program codes
- **FPP_to_School_enrollment**: The ratio of FPP to school enrollment
- **Retained**: target




### Finance
- **ID_FINANCE**: ID finance department
- **Deposit_Date**: Expected deposit day
- **Special_Pay**: Payment modality (internal code)
- **Tuition**: Price per full-payment participant (FPP)
- **FRP_Active**: Number of FPPs who bought trip-cancellation insurance
- **FRP_Cancelled**: Number of FPPs who bought trip-cancellation insurance and cancelled it
- **FRP_Take_up_percent_**: Percentage of FPPs who bought the insurance pay for it
- **EZ_Pay_Take_Up_Rate**: Percentage of FPPs use automatic bank draft
- **School_Sponsor**: Indication of whether or not the school is sponsoring the trip
- **SPR_Group_Revenue**: Amount paid for all of the participants
- **FPP_to_PAX**: Percentage of FPP
- **Num_of_Non_FPP_PAX**: Number of non-FPP participants


### CRM
- **ID_CRM**: ID CRM department
- **Poverty_Code**: Poverty code for the school area based on estimated percentage below the poverty line. A is 0 to 5.9, B is 6 to 15.9, C is 16 to 30.9, D is 31 or more, E is unclassified, Space if DISTCLASS = U (Supervisory Union)
- **Region**: State areas
- **CRM_Segment**: CRM code system (internal code)
- **School_Type**: Public or private
- **Parent_Meeting_Flag**: Indication whether a parent meeting was held
- **MDR_Low_Grade**: Lowest grade (not just participants) in the school
- **MDR_High_Grade**: Highest grade (not just participants) in the school
- **Total_School_Enrollment**: School enrollments
- **Income_Level**: Parent income level code. A is lowest, Q is highest, Z is unclassified
- **SPR_New_Existing**: New client indicator
- **NumberOfMeetingswithParents**: Number of meetings with parents prior to the trip
- **FirstMeeting**: The date of the first meeting with parents (NA if none held)
- **LastMeeting**: Date of the last meeting with parents (NA if none held)
- **DifferenceTraveltoFirstMeeting**: Days from the first parent meeting to travel date
- **DifferenceTraveltoLastMeeting**: Days from the last parent meeting to travel date
- **SchoolGradeTypeLow**: The lowest grade type in the school
- **SchoolGradeTypeHigh**: The highest grade type in the school
- **SchoolGradeType**: Combination of the above denoting the type of school
- **SchoolSizeIndicator**: Size of the school (S, M, L, S-M, M-L)

## Create `ID_TRIP`

`ID_TRIP` is the ID that joins `finance_df` and `sales_df` in a one-to-one manner.

### Finance DF

```python
# Preliminary check
# Check unique values for `Special_Pay` (values used to build the ID) to see if they need special handling

print(f"Unique values for `Special_Pay`: {finance_df['Special_Pay'].unique()}")
```

The only special value for `Special_Pay` is `0` which resolves to an empty string `''`.

Also `NA` can be considered a special value since when importing the CSV through pandas it resolves to `nan` which introduces a problem. We fix this by adding to the import `keep_default_na=False`.

```python
# Creating `ID_TRIP`

def calc_id_trip_finance(row):
  special_pay = row["Special_Pay"] if row["Special_Pay"] != '0' else '' # Ternary check to resolve `0` to an empty string `''`
  ID = row["ID_FINANCE"]
  
  if ID.startswith(special_pay):
    return ID[len(special_pay):]
  else:
    raise ValueError(f"Error: Row ID '{row.name}' => '{ID}' does not start with '{special_pay}.'") # Throw error if not found, in that case ID calculation is wrong
  
finance_df["ID_TRIP"] = finance_df.apply(calc_id_trip_finance, axis=1)

display(finance_df.head())
```

### Sales DF

```python
# Preliminary check
# Check unique values for `Program_Code` and `Travel_Type` (values used to build the ID) to see if they need special handling

print(f"Unique values for `Program_Code`: {sales_df['Program_Code'].unique()}")
print(f"Unique values for `Travel_Type`: {sales_df['Travel_Type'].unique()}")
```

There is no value that requires any particular handling. 

```python
# Creating `ID_TRIP`

def calc_id_trip_sales(row):
  program_code = row['Program_Code']
  travel_type = row['Travel_Type']
  ID = row['ID_SALES']
  
  if ID.startswith(program_code) and ID.endswith(travel_type):
    return ID[len(program_code):-len(travel_type)]
  else:
    # raise ValueError(f"Error: Row ID '{row.name}' => '{ID}' does not start with '{program_code}' AND/OR does not end with '{travel_type}'.") # Throw error if not found, in that case ID calculation is wrong
    print(f"Error: Row ID '{row.name}' => '{ID}' does not start with '{program_code}' AND/OR does not end with '{travel_type}'.") # Throw error if not found, in that case ID calculation is wrong
    # Looks like the 3 not matching because they don't end with a Travel_Type end with a '0'. This makes id 5 digits long which is the only case that happens, so we try and remove the 0.
    return ID[len(program_code):-1] 
  
sales_df["ID_TRIP"] = sales_df.apply(calc_id_trip_sales, axis=1)

display(sales_df.head())
```

### CRM DF

```python
# Preliminary check
# Check unique values for `Poverty_Code` and `Income_Level` (values used to build the ID) to see if they need special handling

print(f"Unique values for `Poverty_Code`: {crm_df['Poverty_Code'].unique()}")
print(f"Unique values for `Income_Level`: {crm_df['Income_Level'].unique()}")
```

There is no value that requires any particular handling. 

```python
# Creating `ID_TRIP`

def calc_id_trip_crm(row):
  poverty_code = row['Poverty_Code']
  income_level = row['Income_Level']
  ID = row['ID_CRM']
  
  if ID.startswith(poverty_code) and ID.endswith(income_level):
    return ID[len(poverty_code):-len(income_level)]
  else:
    raise ValueError(f"Error: Row ID '{row.name}' => '{ID}' does not start with '{poverty_code}' AND/OR does not end with '{income_level}'.") # Throw error if not found, in that case ID calculation is wrong
  
crm_df["ID_TRIP"] = crm_df.apply(calc_id_trip_crm, axis=1)

display(crm_df.head())
```

```python
# Check for duplicates to see if we have identified ID correctly

duplicates = finance_df['ID_TRIP'].value_counts()
duplicates = duplicates[duplicates > 1]  # Filter values that appear more than once

print(len(duplicates))

duplicates = sales_df['ID_TRIP'].value_counts()
duplicates = duplicates[duplicates > 1]  # Filter values that appear more than once

print(len(duplicates))

duplicates = crm_df['ID_TRIP'].value_counts()
duplicates = duplicates[duplicates > 1]  # Filter values that appear more than once

print(len(duplicates))
```

## Merge into `trips_df`

`trips_df` is a df containing all information regarding a trip: the actual trip data (gathered from `sales_df`) and the financial data (gathered from `financial_df`)

```python
# Try merging with 'outer' so we see how many don't get matched

print("Merging Finance DF + Sales DF")

outer_finance_sales_df = finance_df.merge(sales_df, on="ID_TRIP", how="outer", indicator=True)

unmatched_rows = outer_finance_sales_df[outer_finance_sales_df["_merge"] != "both"]

print(f"Finance DF Lenght: {len(finance_df)}")
print(f"Sales DF Lenght: {len(sales_df)}")
print(f"Unmatched: {len(unmatched_rows)}")
display(unmatched_rows[["ID_FINANCE", "ID_SALES"]])

outer_finance_sales_df = outer_finance_sales_df.drop("_merge", axis=1)

# Now merge inner

inner_finance_sales_df = finance_df.merge(sales_df, on="ID_TRIP")

print(f"Merged DF Lenght: {len(inner_finance_sales_df)}")
display(inner_finance_sales_df)

# Try merging with 'outer' so we see how many don't get matched

print("Merging inance DF + Sales DF + CRM DF")

outer_finance_sales_crm_df = outer_finance_sales_df.merge(crm_df, on="ID_TRIP", how="outer", indicator=True)
unmatched_rows = outer_finance_sales_crm_df[outer_finance_sales_crm_df["_merge"] != "both"]

print(f"Finance DF Lenght: {len(finance_df)}")
print(f"Sales DF Lenght: {len(sales_df)}")
print(f"CRM DF Lenght: {len(crm_df)}")
print(f"Unmatched: {len(unmatched_rows)}")
display(unmatched_rows[["ID_FINANCE", "ID_SALES", "ID_CRM"]])

# Now merge inner

inner_finance_sales_crm_df = inner_finance_sales_df.merge(crm_df, on="ID_TRIP")

print(f"Merged DF Lenght: {len(inner_finance_sales_crm_df)}")
display(inner_finance_sales_crm_df)
```
